**Data Transformations**
- Generally data has 3 types of distributions
  - Normal distribution
  - Positive skewed
  - Negative skewed

- All the math concepts developed by assumption as data follows normal distribution
- Data skew happens because of many reasons, in that one reason is outliers
- once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normally
- We have some transfomation techniques
  - Log transformation
  - Exponential transformation
  - Reciprocal transformation
  - Power transformation
  - Square root transformation

**Log Transformation**
- Log transformation means performing logarithm operations on original data
- It is an approach to convert into normal distribution
- Log means natural logarithm base=e

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=20,label="Exponential")
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=20,label="Logritham")  #bins means intervals
plt.legend()
plt.show()
```

**Squar root**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label="Square_root_data")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1).hist(exp_data,bins=20,label="Exponential")
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label="Logritham")
plt.legend()
plt.subplot(1,3,3).hist(square_root_data,bins=20,label="Square_root_data")
plt.legend()
plt.show()
```

**Power Transformation**

- Box cox
- Ye-jhonson
- it is under sklearn preprocessing

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans = power.fit_transform(exp_data1)
plt.hist(data_trans)
```

```python
from sklearn.preprocessing import PowerTransformer
power_box = PowerTransformer(method='box-cox')
exp_data2=exp_data.reshape(-1,1)
data_trans_box= power_box.fit_transform(exp_data2)
plt.hist(data_trans_box)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(data_trans,bins=20,label='yeo-johnson')
plt.legend()
plt.subplot(1,2,2).hist(data_trans_box,bins=20,label='box-cox')
plt.legend()
plt.show()
```
