**one hot encoder**

- One hot means if one is ON another one is OFF
- ON represents with 1 and OFF reprsents with 0
- For example, case status has two unique lables
    - Denied
    - Certified
- if we apply one hot encoder on case status column it will create two extra columns
    - case_status_Denied
    - case_status_Certified

```python
case_status     case_status_denied    case_status_certfied
 Denied           1                       0
 Certified        0                       1
```

**Advantages**
- In Machine learning we have one important property is **Independent of variables**
- Machine learning models expect variables are **independent each other**
- One hot encoder creates new columns, the new columns are **independent each otehr**
- Independet variables means both variables have **90 degrees phase shift**
- **90 degrees phase shift** also called as **perpanducular each other**
- **perpanducular each other** means **orthogonal each other**

**Disadvantages**
- One hot econder creates new columns equal to number of unique lables
- if number of unique lables increase, new columns also increase
- The more columns leads data is more
- then data is more the complexity will increase
- Complexity increases more computation power required
- Also time increase
- The entire process name as **curse of dimensionality**
