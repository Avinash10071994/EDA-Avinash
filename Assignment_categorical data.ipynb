```python
unique_cnt=visa_df["education_of_employee"].unique()
count_list=[]
for i in unique_cnt:
    con=visa_df["education_of_employee"]==i
    val=len(visa_df[con])
    count_list.append(val)
    print(i,val)
```

```python
employee_df= pd.DataFrame(zip(unique_cnt,count_list),
             columns=["education_of_employee","count"])
employee_df
```

```python
plt.bar("education_of_employee","count",data=employee_df)
plt.title("Barchart")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.savefig(r"C:\Users\avina\OneDrive\Documents\NareshIT\DATA SCIENCE & AI\EDA Python\Barchart_education_of_employee.jpg")
plt.show()
```
